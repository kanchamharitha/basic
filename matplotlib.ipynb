```python
import matplotlib.pyplot as plt
```

```python
#1. Create a line chart showing Month vs Sales.
x=df["Month"]
y=df["Sales"]
plt.plot(x,y,marker="*",color="black")
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Trends")
plt.show()
```

```python
#3. Plot Sales and Profit on the same chart.

x=df["Month"]
y=df["Profit"]
y1=df["Sales"]
plt.plot(x,y,marker="*",color="black")
plt.plot(x,y1,marker="o",color="green")
plt.xlabel("Month")
plt.ylabel("profit,sales")
plt.title("Trends",color="blue",fontsize=20)
plt.show()
```

```python
#2. Create a line chart showing Month vs Profit.
x=df["Month"]
y=df["Profit"]
plt.plot(x,y,marker="*",color="black")
plt.xlabel("Month")
plt.ylabel("profit")
plt.title("Trends")
plt.show()
```

```python
#4. Add markers to the Sales line chart.
x=df["Month"]
y=df["Profit"]
plt.plot(x,y,marker="s",color="black",markerfacecolor="red")
plt.xlabel("Month")
plt.ylabel("profit")
plt.title("Trends")
plt.show()
```

```python
#5. Add a grid to the line chart.
x=df["Month"]
y=df["Profit"]
plt.plot(x,y,marker="s",color="black",markerfacecolor="red")
plt.xlabel("Month")
plt.ylabel("profit")
plt.title("Trends")
plt.grid()
plt.show()
```

```python
#6. Create a bar chart showing Month-wise Sales.
x=df["Month"]
y=df["Sales"]
plt.bar(x,y)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.title("Trends")
plt.show()
```

```python
#7. Create a bar chart showing Month-wise Profit.
x=df["Month"]
y=df["Profit"]
plt.bar(x,y)
plt.xlabel("Month")
plt.ylabel("profit")
plt.title("Trends")
plt.show()

```
