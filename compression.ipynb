```python
or_lst=[[10,9],[6,7],9,0]
sh_lst=or_lst.copy()
print(or_lst)
print(sh_lst)
```

```python
or_lst[0][1]=11
print(or_lst)
print(sh_lst)
```

```python
import copy
de_lst=copy.deepcopy(or_lst)
print(or_lst)
print(de_lst)
```

```python
de_lst[0][1]=6
print(or_lst)
print(de_lst)
```

```python

de_lst[0][1]=6
print("deepcopy list:",de_lst)
sh_lst=or_lst.copy()
print("shallow list:",sh_lst)
```
