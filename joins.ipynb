```python
#inner join
import pandas as pd
df1=pd.DataFrame({"ID":[1,2,3,4,5],"Name":["A","B","C","D","E"]})
df2=pd.DataFrame({"ID":[2,4,5,6,7],"City":["Ap","kerala","chennai","mumbai","delhi"]})
result=pd.merge(df1,df2,on="ID",how="inner")
print(result)
```

```python
#left outer join
result=pd.merge(df1,df2,on="ID",how="left")
print(result)
```

```python
#right outer join
result=pd.merge(df1,df2,on="ID",how="right")
print(result)
```

```python
#cross join
result=df1.merge(df2,how="cross")
print(result)
```

```python
#self join
df1=pd.DataFrame({"ID":[1,2,3,4,5],"Name":["A","B","C","D","E"],"Salary":[100,120,200,400,500]})
result=df1.merge(df1,left_on="Salary",right_on="ID",how="inner",suffixes=('(e)','(m)'))
print(result)
```

```python
df=pd.read_csv("/content/Customers.csv")
df.head()
```

```python
pro=pd.read_csv("/content/Orders.csv")
pro.head()
```

```python
#1. Display all customer details along with their orders using an Inner Join.
result=pd.merge(df,pro,on="CustomerID",how="inner")
print(result)
```

```python
#2. Show only those customers who have placed at least one order.
result=pd.merge(df,pro,on="CustomerID",how="inner")
print(result)
```

```python
#3. Display all customers even if they have not placed any orders (Left Join).
result=pd.merge(df,pro,on="CustomerID",how="left")
print(result)
```

```python
#4. Find customers who have never placed an order.
customers_no_orders = df[df['CustomerID'].isna()][['CustomerID', 'CustomerName', 'City']]
print(customers_no_orders)
```

```python
#5. Display all orders even if customer information is missing (Right Join).
result=pd.merge(df,pro,on="CustomerID",how="right")
print(result)
```

```python
#6. Find orders whose CustomerID does not exist in the Customers table.
result=pd.merge(df,pro,on="CustomerID",how="outer")
print(result)
```

```python
#7. Perform a Full Outer Join between Customers and Orders.
result=pd.merge(df,pro,on="CustomerID",how="outer")
print(result)
```

```python
#8. Identify records that exist only in Customers but not in Orders.
result=pd.merge(df,pro,on="CustomerID",how="left")
print(result)
```

```python
#9. Identify records that exist only in Orders but not in Customers.
result=pd.merge(df,pro,on="CustomerID",how="right")
print(result)
```

```python
#10. Merge Orders and Payments using OrderID.
pay=pd.read_csv("/content/Payments.csv")
pay.head()
result=pd.merge(pro,pay)
print(result)
```

```python
#11. Find orders for which payment information is not available.
result=pd.merge(pro,pay,on="OrderID",how="outer")
print(result)
```
