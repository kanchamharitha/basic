```python
#56. Find row-wise sum.
row_sum = []
for i in range(arr.shape[0]):   # iterate over rows
    total = 0
    for j in range(arr.shape[1]):  # iterate over columns
        total += arr[i][j]
    row_sum.append(total)

print(row_sum)
```
