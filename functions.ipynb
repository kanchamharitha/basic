```python
#Create a function to calculate the factorial of a number.
def fact(a):
  k=1
  for i in range(1,a+1):
    k=k*i
  return k
fact(5)
```

```python
#Create a function to count vowels in a string.
def count(s):
  count=0
  for i in s:
    if i in "aeiou":
      count+=1
  return count
count("korea")
```

```python
#Create a function to reverse a string.
def reverse(s):
  return s[ : :-1]
reverse("mogli")


```

```python
#Create a function to check whether a string is palindrome or not.
def palindrom(g):
  t=g[: :-1]
  if g==t:
    print("palindrom")
  else:
    print("not a palindrom")
  return
palindrom("madam")
```

```python
#41. Create a function using *args to find sum of numbers.
def sum(*args):
  t=0
  for i in args:
    t+=i
  return t
sum(1,2,3,4,5)
```

```python
#42. Create a function using *args to find maximum number.
from functools import reduce
def sum(*args):

  return reduce(lambda a,b:a if a>b else b,args)


sum(1,2,3,4,5)
```
