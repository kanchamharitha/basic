```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
df=pd.read_csv("/content/tips.csv")
df.head()
```

```python
#scatterplot
sns.scatterplot(
    data=df,
    x="total_bill",
    y="tip",
    hue="sex",
    size="size",
    sizes=(50,300),
    marker="o",
    alpha=0.9,
    palette="Accent"
)
plt.show()
```

```python
#linechart
sns.lineplot(
    data=df,
    x="size",
    y="total_bill",
    hue="sex",
    style="smoker",
    marker=True,
    alpha=0.9,
    palette="Set1"
)
plt.show()
```

```python
#bar plot
sns.barplot(
    data=df,
    x="size",
    y="total_bill",
    hue="sex",


    alpha=0.9,
    palette="Set1"
)
plt.show()
```

```python
#Histogram
sns.histplot(
    data=df,
    x="total_bill",
    bins=10,
    hue="sex",
    alpha=0.9,
    palette="Set2"
)
plt.show()
```

```python
#KDE plt
sns.kdeplot(
    data=df,
    x="total_bill",
    hue="smoker",
    linewidth=3,
    fill=True,
    palette="Set3"

)
plt.show()
```

```python
#violin plot
sns.violinplot(
    data=df,
    x="day",
    y="total_bill",
    hue="time",
    palette="Accent",
    inner="quartile"
)
plt.show()
```

```python
#heat map
corr=df.corr(numeric_only=True)
print(corr)
```

```python
sns.heatmap(
    corr,
    cmap="RdBu"

)
plt.show()
```

```python
sns.heatmap(
    corr,
    cmap="PuOr"

)
plt.show()
```

```python
#catplot
sns.catplot(
    data=df,
    x="day",
    y="total_bill",
    kind="boxen",
    hue="sex"
)
plt.show
```
