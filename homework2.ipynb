```python
#Phase 1: Data Cleaning
#What is the shape of the dataset?
import pandas as pd
df=pd.read_csv("/content/dataset.csv")
df.shape
```

```python
#Which columns contain missing values?
#How many missing values are present in each column?

df.isnull().sum()
```

```python
#What percentage of missing values is present in each column?
df.isnull().sum()/len(df)*100
```

```python
#How will you handle the missing values?
df["artists"].fillna(df["artists"].mode()[0],inplace=True)
df["album_name"].fillna(df["album_name"].mode()[0],inplace=True)
df["track_name"].fillna(df["track_name"].mode()[0],inplace=True)
```

```python
df.isnull().sum()
```

```python
#Are there any duplicate records in the dataset?
df.duplicated()
```

```python
#How many duplicate records are present?
df.duplicated().sum()
```

```python
#Remove duplicate records and verify the result.
df.duplicated(subset=["Unnamed: 0"]).sum()
```

```python
#Is the Unnamed: 0 column useful for analysis?
df['Unnamed: 0']
df.loc[0:5]
```

```python
#Remove unnecessary columns from the dataset.
df.drop(columns=["Unnamed: 0"],axis=1,inplace=True)
```
