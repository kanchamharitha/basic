```python
#Linear Regression
import pandas as pd
from sklearn.linear_model import LinearRegression
df=pd.read_csv("tips.csv")
df.head()
```

```python
x=df[["tip"]]
y=df["total_bill"]
lr=LinearRegression()
print(lr.fit(x,y))
```

```python
print(lr.coef_)
print(lr.intercept_)
print(lr.predict([[4]]))
```

```python
#multilinear Regression
x=df[["tip","size"]]
y=df["total_bill"]
print(lr.fit(x,y))
```

```python
print(lr.coef_)
print(lr.intercept_)
print(lr.predict([[4,8]]))
```

```python
#training and testing data
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error,mean_squared_error,r2_score
x=df[["tip"]]
y=df["total_bill"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
lr.fit(x_train,y_train)
```

```python
y_test_pred=lr.predict(x_test)
y_test_pred
```

```python
y_test
```

```python
y_train_pred=lr.predict(x_train)
y_train_pred
```

```python
accuracy=r2_score(y_test_pred,y_test)
accuracy
```

```python
accuracy_train=r2_score(y_train_pred,y_train)
accuracy_train
```

```python
print(len(y_train),len(y_train_pred))
```

```python
mean_absolute_error(y_test_pred,y_test)
```

```python
mean_absolute_error(y_train_pred,y_train)
```
