```python
#21. Create decorator to print welcome message before function execution.
def dec_greet(func):
  def wrapper():
    print("Good Morning")
    func()
  return wrapper
@dec_greet
def greet():
  print("hello students")
greet()
```

```python
#22. Create decorator to print before and after function execution.
def dec_before(func):
  def wrapper():
    print("Good Morning")
    func()
  return wrapper
@dec_before
def greet():
  print(" thank you everyone")
greet()
```

```python
#24. Create decorator to calculate function execution time.
def dec_execution(func):
  def wrapper():
    print("execution is not done")
    func()
  return wrapper
@dec_execution
def greet():
  print("execution is done")
greet()
```

```python
#25. Create decorator to convert output into uppercase.
def dec_execution(func):
  def wrapper(i):
    print(i)
    func(i)
  return wrapper
@dec_execution
def greet(i):
  print(i.upper())
i="archana"
greet(i)
```

```python
#30. Create decorator to check integer input before calculation.
def dec_login(func):
  def wrapper(a,b):
    print(f"the addition of {a} and {b} is:")
    func(a,b)
  return wrapper
@dec_login
def greet(x,y):
  print(x+y)
greet(3,8)
```
